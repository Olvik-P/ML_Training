# Что такое глубокое обучение? #

Одними из самых впечатляющих достижений в области искусственного интеллекта за последние годы стали успехи в сфере *глубокого обучения*. Машинный перевод, распознавание изображений и игры — всё это задачи, в которых модели глубокого обучения приблизились или даже превзошли человеческий уровень.

Так что же такое глубокое обучение? **Глубокое обучение** — это подход к машинному обучению, характеризующийся глубокими стеками вычислений. Именно эта глубина вычислений позволяет моделям глубокого обучения распутывать сложные и иерархические закономерности, встречающиеся в самых сложных реальных наборах данных.

Благодаря своей мощности и масштабируемости **нейронные сети** стали определяющей моделью глубокого обучения. Нейронные сети состоят из нейронов, каждый из которых выполняет лишь простое вычисление. Сила нейронной сети заключается в сложности связей, которые эти нейроны могут образовывать.

# Линейный нейрон #

Итак, начнём с фундаментального компонента нейронной сети — отдельного нейрона. В виде диаграммы **нейрон** (или **узел**) с одним входом выглядит так:

<figure style="padding: 1em;">
<img src="img/mfOlDR6.png" width="250" alt="Диаграмма линейного нейрона.">
<figcaption style="textalign: center; font-style: italic"><center>Линейный нейрон: $y = w x + b$
</center></figcaption>
</figure>

Вход — это `x`. Его соединение с нейроном имеет **вес** `w`. Когда значение проходит через соединение, оно умножается на вес этого соединения. Для входа `x` до нейрона доходит `w * x`. Нейронная сеть «обучается», изменяя свои веса.

`b` — это особый вид веса, который мы называем **смещением** (bias). Смещение не связано с какими-либо входными данными; вместо этого на диаграмме мы ставим `1`, так что до нейрона доходит просто `b` (поскольку `1 * b = b`). Смещение позволяет нейрону изменять выход независимо от входов.

`y` — это значение, которое нейрон в итоге выдаёт. Чтобы получить выход, нейрон суммирует все значения, полученные через свои соединения. Активация этого нейрона равна `y = w * x + b$, или в виде формулы $y = w x + b$.

<blockquote style="margin-right:auto; margin-left:auto; background-color: #ebf9ff; padding: 1em; margin:24px;">
    <strong>Формула $y = w x + b$ кажется знакомой?</strong><br>
Это уравнение прямой! Это уравнение с угловым коэффициентом, где $w$ — наклон, а $b$ — точка пересечения с осью $y$.
</blockquote>

# Пример — линейный нейрон как модель #

Хотя отдельные нейроны обычно функционируют лишь как часть более крупной сети, часто полезно начать с модели из одного нейрона в качестве базового уровня. Модели из одного нейрона являются *линейными* моделями.

Давайте подумаем, как это может работать на наборе данных [80 Cereals](https://www.kaggle.com/crawford/80-cereals). Обучив модель с `'sugars'` (граммы сахара на порцию) на входе и `'calories'` (калории на порцию) на выходе, мы можем получить смещение `b = 90` и вес `w = 2,5`. Мы можем оценить калорийность хлопьев с 5 граммами сахара на порцию следующим образом:

<figure style="padding: 1em;">
<img src="img/yjsfFvY.png" width="1000" alt="Вычисление с помощью линейного нейрона.">
<figcaption style="textalign: center; font-style: italic"><center>Вычисление с помощью линейного нейрона.
</center></figcaption>
</figure>

И, сверяясь с нашей формулой, получаем $calories = 2,5 \times 5 + 90 = 102,5$, как и ожидалось.

# Несколько входов #

Набор данных *80 Cereals* содержит гораздо больше признаков, чем просто `'sugars'`. Что, если мы захотим расширить нашу модель, включив в неё, например, содержание клетчатки или белка? Это несложно. Мы можем просто добавить больше входных соединений к нейрону — по одному на каждый дополнительный признак. Чтобы получить выход, мы умножим каждый вход на его вес и затем сложим всё вместе.

<figure style="padding: 1em;">
<img src="img/vyXSnlZ.png" width="300" alt="Три входных соединения: x0, x1 и x2, а также смещение.">
<figcaption style="textalign: center; font-style: italic"><center>Линейный нейрон с тремя входами.
</center></figcaption>
</figure>

Формула для этого нейрона будет выглядеть так: $y = w_0 x_0 + w_1 x_1 + w_2 x_2 + b$. Линейный нейрон с двумя входами будет задавать плоскость, а нейрон с бóльшим количеством входов — гиперплоскость.

# Линейные нейроны в Keras #

Самый простой способ создать модель в Keras — использовать `keras.Sequential`, который создаёт нейронную сеть как стек *слоёв*. Мы можем создавать модели, подобные описанным выше, с помощью *полносвязного* (dense) слоя (подробнее о нём мы узнаем на следующем уроке).

Мы могли бы определить линейную модель, принимающую три входных признака (`'sugars'`, `'fiber'` и `'protein'`) и выдающую один выход (`'calories'`), следующим образом:

```python
python -m pip install tensorflow
```

```python
from tensorflow import keras
from tensorflow.keras import layers

# Создаём сеть с 1 линейным нейроном
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[3])
])
```

Первым аргументом `units` мы задаём, сколько выходов нам нужно. В данном случае мы прогнозируем только `'calories'`, поэтому используем `units=1`.

Вторым аргументом `input_shape` мы сообщаем Keras размерность входных данных. Установка `input_shape=[3]` гарантирует, что модель будет принимать три признака на входе (`'sugars'`, `'fiber'` и `'protein'`).

Теперь эта модель готова к обучению на данных!

<blockquote style="margin-right:auto; margin-left:auto; background-color: #ebf9ff; padding: 1em; margin:24px;">
    <strong>Почему <code>input_shape</code> — это список Python?</strong><br>
Данные, которые мы будем использовать в этом курсе, будут табличными, как в DataFrame Pandas. У нас будет один вход для каждого признака в наборе данных. Признаки расположены по столбцам, поэтому мы всегда будем использовать <code>input_shape=[num_columns]</code>.

Причина, по которой Keras использует здесь список, — возможность работы с более сложными наборами данных. Например, для изображений могут потребоваться три измерения: <code>[height, width, channels]</code>.
</blockquote>
