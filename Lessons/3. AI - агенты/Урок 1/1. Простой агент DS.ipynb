## 📖 Теория на пальцах

### Что такое AI-агент?

Представьте **агента** как робота-пылесоса:
1. **Восприятие** — видит датчиками: «впереди стена?» или «тут мусор?»
2. **Анализ** — решает по правилу: «Если вижу мусор → убирать. Если стена → повернуть»
3. **Действие** — едет, поворачивает, собирает пыль

**Формула агента:** `Правила (если → то) + Память (словарь) + Цикл (постоянно работает)`

Сегодня мы сделаем **агента-термостата** — он живёт в комнате и решает: включить обогрев или нет.

---
### Среда, где живёт агент

**Среда** — это мир вокруг агента. У нашего агента среда — это просто число (температура в комнате).

Агент может:
- **воспринимать** температуру
- **действовать** (включить/выключить нагреватель)

Никакой магии. Всё как в жизни.

## 🌡️ Пример: агент-термостат (полностью готовый)

Этот агент следит за температурой. Если холодно — включает обогрев. Если жарко — выключает. Проще некуда.

```python
# АГЕНТ-ТЕРМОСТАТ (полностью рабочий пример)

class ThermostatAgent:
    def __init__(self, target_temp=22):
        self.target_temp = target_temp  # комфортная температура
        self.heater_on = False          # состояние обогревателя
        print(f"🌡️ Агент создан. Цель: {self.target_temp}°C")
    
    def perceive(self, current_temp):
        """Шаг 1: Восприятие — агент узнаёт температуру"""
        self.current_temp = current_temp
        return self.current_temp
    
    def decide(self):
        """Шаг 2: Анализ — принимает решение по правилам"""
        if self.current_temp < self.target_temp - 1:
            return "HEAT_ON"     # слишком холодно → включить
        elif self.current_temp > self.target_temp + 1:
            return "HEAT_OFF"    # слишком жарко → выключить
        else:
            return "DO_NOTHING"  # нормально
    
    def act(self, decision):
        """Шаг 3: Действие — выполняет решение"""
        if decision == "HEAT_ON" and not self.heater_on:
            self.heater_on = True
            return "🔥 Включаю обогрев"
        elif decision == "HEAT_OFF" and self.heater_on:
            self.heater_on = False
            return "❄️ Выключаю обогрев"
        elif decision == "DO_NOTHING":
            return "😌 Всё хорошо, ничего не делаю"
        else:
            return "⚙️ Без изменений"
    
    def step(self, current_temp):
        """Один полный цикл агента: восприятие → анализ → действие"""
        self.perceive(current_temp)
        decision = self.decide()
        action_message = self.act(decision)
        return f"Температура: {current_temp}°C → {action_message}"

# ------------------- ЗАПУСКАЕМ АГЕНТА -------------------
print("\n" + "="*50)
agent = ThermostatAgent(target_temp=22)

# Симулируем изменение температуры в комнате
test_temperatures = [18, 20, 22, 23, 21, 15, 25, 22]

print("\n🏠 Комната начинает остывать и нагреваться...\n")
for temp in test_temperatures:
    result = agent.step(temp)
    print(result)
    time.sleep(0.5)  # маленькая пауза, чтобы было красиво

print("\n✅ Агент отработал все циклы. Видите? Простые правила → умное поведение.")
```

## 🧠 Как это работает? (разбор по шагам)

1. **`perceive`** — агент получает температуру из внешнего мира
2. **`decide`** — смотрит на правила: `если холоднее чем нужно → HEAT_ON`
3. **`act`** — меняет состояние обогревателя и сообщает об этом

Весь смысл AI-агента — **цикл**: восприятие → решение → действие → снова восприятие.

💡 В реальных роботах этот цикл повторяется тысячи раз в секунду.

## 🧪 Практическое задание: написать своего агента «Пылесос в комнате»

**Ваша задача:** создать агента, который живёт в комнате 3×3 (9 клеток).

**Правила среды:**
- В комнате в случайных клетках есть мусор (⭐)
- Агент стоит в клетке (0,0) и видит: есть ли мусор в текущей клетке

**Что должен делать агент:**
1. Если в текущей клетке есть мусор → убрать (действие `CLEAN`)
2. Если мусора нет → переместиться в случайную соседнюю клетку (действие `MOVE`)
3. Агент работает 10 шагов, а потом сообщает, сколько мусора убрал

💡 **Подсказка:** сначала допишите код в следующей ячейке, используя шаблон.

```python
# 🧹 АГЕНТ-ПЫЛЕСОС — ваша задача дописать

def new_coordinat(val: int) -> int:
    if val == 0:
        step = val + random.choice([0, 1])
    elif val == 2:
        step = val - random.choice([0, 1])
    else:
        step = val + random.choice([-1, 0, 1])
    return step

class VacuumAgent:
    def __init__(self, room):
        """
        room: двумерный список (3x3), где 1 = мусор, 0 = чисто
        """
        self.room = room
        self.row = 0      # текущая позиция агента (строка)
        self.col = 0      # текущая позиция агента (столбец)
        self.cleaned = 0  # счётчик убранного мусора
        print("🤖 Пылесос создан! Начинаю уборку...")
    
    def perceive(self):
        """Восприятие: есть ли мусор в текущей клетке?"""
        # ДОПИШИ: проверь self.room[self.row][self.col], верни True/False
        return self.room[self.row][self.col]
    
    def decide(self, has_trash):
        """Анализ: если есть мусор → CLEAN, иначе → MOVE"""
        # ДОПИШИ: если has_trash True, верни "CLEAN", иначе "MOVE"
        return "CLEAN" if has_trash else "MOVE"
    
    def _new_coordinat(self, val: int) -> int:
        if val == 0:
            return val + random.choice([0, 1])
        elif val == len(self.room) - 1:
            return val - random.choice([0, 1])
        else:
            return val + random.choice([-1, 0, 1])

    def act(self, decision):
        """Действие: убирает мусор или двигается"""
        if decision == "CLEAN":
            # Убираем мусор (ставим 0 в комнате)
            self.room[self.row][self.col] = 0
            self.cleaned += 1
            return f"🧹 Убрал мусор в клетке ({self.row},{self.col})"
        else:  # MOVE
            self.row = self._new_coordinat(self.row)
            self.col = self._new_coordinat(self.col)
            return f"🚶‍♂️ Переместился в ({self.row},{self.col})"
    
    def step(self):
        """Один цикл агента"""
        trash_here = self.perceive()
        action = self.decide(trash_here)
        message = self.act(action)
        return message
    
    def run(self, steps=10):
        """Запускаем агента на несколько шагов"""
        print("="*50)
        for i in range(steps):
            print(f"Шаг {i+1}: {self.step()}")
            time.sleep(0.3)
        print(f"\n✅ Уборка завершена. Убрано {self.cleaned} единиц мусора.")
        return self.cleaned

# ------------------- СОЗДАЁМ КОМНАТУ -------------------
random.seed(42)  # чтобы результат был одинаковым у всех (для проверки)
room_with_trash = [[random.choice([0,1]) for _ in range(3)] for _ in range(3)]
print("🗺️ Карта комнаты (1 = мусор, 0 = чисто):")
for row in room_with_trash:
    print(row)

# ------------------- ЗАПУСКАЕМ ВАШЕГО АГЕНТА -------------------
agent = VacuumAgent(room_with_trash)
cleaned_total = agent.run(steps=10)
```

## ✅ Проверь себя (тесты для самопроверки)

Запустите ячейку ниже. Если всё сделали правильно — ошибок не будет.

```python
# ТЕСТЫ для вашего агента-пылесоса

# Тест 1: правильное восприятие
test_room = [[1,0,0],[0,0,0],[0,0,0]]
test_agent = VacuumAgent(test_room)
assert test_agent.perceive() == True, "❌ В клетке (0,0) есть мусор (1), а perceive вернул не True"
test_agent.act("CLEAN")
assert test_agent.room[0][0] == 0, "❌ act('CLEAN') не убрал мусор (должен был поставить 0)"
print("✅ Тест 1 пройден: восприятие и уборка работают")

# Тест 2: решение (decide)
assert test_agent.decide(True) == "CLEAN", "❌ Если есть мусор, надо вернуть 'CLEAN'"
assert test_agent.decide(False) == "MOVE", "❌ Если мусора нет, надо вернуть 'MOVE'"
print("✅ Тест 2 пройден: логика принятия решений правильная")

# Тест 3: счётчик убранного мусора
assert test_agent.cleaned == 1, "❌ Агент убрал 1 мусор, но cleaned = " + str(test_agent.cleaned)
print("✅ Тест 3 пройден: счётчик работает")

print("\n🎉 Отлично! Ваш агент проходит все тесты!")
print("💡 Вы только что создали настоящего AI-агента. Без нейросетей, без API — на чистом Python.")
```
