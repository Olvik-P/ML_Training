## 📖 Теория на пальцах

### Что такое AI-агент?

Представьте **агента** как робота-пылесоса:
1. **Восприятие** — видит датчиками: «впереди стена?» или «тут мусор?»
2. **Анализ** — решает по правилу: «Если вижу мусор → убирать. Если стена → повернуть»
3. **Действие** — едет, поворачивает, собирает пыль

**Формула агента:**
```
Правила (если → то) + Память (словарь/переменные) + Цикл (постоянно работает)
```

Сегодня мы сделаем **агента-термостата** — он живёт в комнате и решает: включить обогрев или нет.

---
### Среда, где живёт агент

**Среда** — это мир вокруг агента. У нашего агента среда — это просто число (температура в комнате).

Агент может:
- **воспринимать** температуру
- **действовать** (включить/выключить нагреватель)

Никакой магии. Всё как в жизни.

💡 **Запомните:** агент = цикл «получил данные → принял решение → сделал действие».

## 🌡️ Пример: агент-термостат (полностью готовый)

Этот агент следит за температурой. Если холодно — включает обогрев. Если жарко — выключает. Проще некуда.

Запустите ячейку и посмотрите, как агент принимает решения.

````python
# АГЕНТ-ТЕРМОСТАТ (полностью рабочий пример)

class ThermostatAgent:
    def __init__(self, target_temp=22):
        self.target_temp = target_temp  # комфортная температура
        self.heater_on = False          # состояние обогревателя
        print(f"🌡️ Агент создан. Цель: {self.target_temp}°C")
    
    def perceive(self, current_temp):
        """Шаг 1: Восприятие — агент узнаёт температуру"""
        self.current_temp = current_temp
        return self.current_temp
    
    def decide(self):
        """Шаг 2: Анализ — принимает решение по правилам"""
        if self.current_temp < self.target_temp - 1:
            return "HEAT_ON"     # слишком холодно → включить
        elif self.current_temp > self.target_temp + 1:
            return "HEAT_OFF"    # слишком жарко → выключить
        else:
            return "DO_NOTHING"  # нормально
    
    def act(self, decision):
        """Шаг 3: Действие — выполняет решение"""
        if decision == "HEAT_ON" and not self.heater_on:
            self.heater_on = True
            return "🔥 Включаю обогрев"
        elif decision == "HEAT_OFF" and self.heater_on:
            self.heater_on = False
            return "❄️ Выключаю обогрев"
        elif decision == "DO_NOTHING":
            return "😌 Всё хорошо, ничего не делаю"
        else:
            return "⚙️ Без изменений"
    
    def step(self, current_temp):
        """Один полный цикл агента: восприятие → анализ → действие"""
        self.perceive(current_temp)
        decision = self.decide()
        action_message = self.act(decision)
        return f"Температура: {current_temp}°C → {action_message}"

# ------------------- ЗАПУСКАЕМ АГЕНТА -------------------
print("\n" + "="*50)
agent = ThermostatAgent(target_temp=22)

# Симулируем изменение температуры в комнате
test_temperatures = [18, 20, 22, 23, 21, 15, 25, 22]

print("\n🏠 Комната начинает остывать и нагреваться...\n")
for temp in test_temperatures:
    result = agent.step(temp)
    print(result)
    time.sleep(0.5)  # маленькая пауза, чтобы было красиво

print("\n✅ Агент отработал все циклы. Видите? Простые правила → умное поведение.")
````

## 🧠 Как это работает? (разбор по шагам)

1. **`perceive`** — агент получает температуру из внешнего мира
2. **`decide`** — смотрит на правила: `если холоднее чем нужно → HEAT_ON`
3. **`act`** — меняет состояние обогревателя и сообщает об этом

Весь смысл AI-агента — **цикл**: восприятие → решение → действие → снова восприятие.

💡 В реальных роботах этот цикл повторяется тысячи раз в секунду.

````python
# ЭКСПЕРИМЕНТ 1: большая комната 5×5
big_room = [[random.choice([0,1]) for _ in range(5)] for _ in range(5)]
print("🧪 Эксперимент: комната 5×5, случайный мусор")
big_agent = VacuumAgent(big_room)

# ДОПОЛНИТЕЛЬНО: добавим метод show_map, чтобы видеть комнату
def show_map(agent):
    print("Текущая карта (🤖 = агент, ⭐ = мусор):")
    for r in range(len(agent.room)):
        row_str = ""
        for c in range(len(agent.room[0])):
            if r == agent.row and c == agent.col:
                row_str += "🤖 "
            elif agent.room[r][c] == 1:
                row_str += "⭐ "
            else:
                row_str += "⬜ "
        print(row_str)
    print()

# Запускаем агента с визуализацией
big_agent.run(steps=15)
show_map(big_agent)

print("\n💡 Попробуйте изменить:")
print("  - количество шагов в run(steps=...)")
print("  - размер комнаты (3 на 4 и т.д.)")
print("  - добавить правило 'если убрал 3 мусора → остановиться'")
````

````python
# ЭКСПЕРИМЕНТ 2: меняем поведение агента
# Попробуйте изменить логику: пусть агент сначала убирает ВСЁ в одной клетке,
# а потом двигается. Или наоборот — пусть двигается, пока не найдёт мусор.

class SmartVacuumAgent(VacuumAgent):
    """Умный пылесос: запоминает, где уже был, и не ходит туда дважды"""
    def __init__(self, room):
        super().__init__(room)
        self.visited = set()  # множество посещённых клеток
        self.visited.add((0, 0))
    
    def act(self, decision):
        if decision == "CLEAN":
            self.room[self.row][self.col] = 0
            self.cleaned += 1
            return f"🧹 Убрал мусор в клетке ({self.row},{self.col})"
        else:
            # Ищем непосещённую соседнюю клетку
            directions = ["up", "down", "left", "right"]
            random.shuffle(directions)
            for d in directions:
                new_row, new_col = self.row, self.col
                if d == "up" and self.row > 0:
                    new_row -= 1
                elif d == "down" and self.row < len(self.room) - 1:
                    new_row += 1
                elif d == "left" and self.col > 0:
                    new_col -= 1
                elif d == "right" and self.col < len(self.room[0]) - 1:
                    new_col += 1
                else:
                    continue
                
                if (new_row, new_col) not in self.visited:
                    self.row, self.col = new_row, new_col
                    self.visited.add((new_row, new_col))
                    return f"🚶‍♂️ Переместился в новую клетку ({self.row},{self.col})"
            
            # Если все соседние уже посещены — идём куда угодно
            if d := random.choice(directions):
                if d == "up" and self.row > 0:
                    self.row -= 1
                elif d == "down" and self.row < len(self.room) - 1:
                    self.row += 1
                elif d == "left" and self.col > 0:
                    self.col -= 1
                elif d == "right" and self.col < len(self.room[0]) - 1:
                    self.col += 1
            return f"🚶‍♂️ Переместился в ({self.row},{self.col}) (уже был тут)"

# Запускаем умного пылесоса
print("🧪 Эксперимент 2: умный пылесос с памятью")
smart_room = [[random.choice([0,1]) for _ in range(3)] for _ in range(3)]
smart_agent = SmartVacuumAgent(smart_room)
smart_agent.run(steps=12)
show_map(smart_agent)
print("💡 Видите разницу? Агент с памятью избегает уже посещённых клеток!")
````
